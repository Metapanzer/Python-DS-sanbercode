```python
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
```

```python

df = sns.load_dataset('diamonds')
display(df.head(), df.tail())
```

# Soal 1: Scatter Plot

Buatlah scatterplot menggunakan Seaborn dari data df di atas, variasikan menggunakan palette yang telah disediakan sehingga didapatkan output sesuai expected.

```python
sns.set({'figure.figsize':(9,9)})
palette = "ch:r=-.2,d=.3_r"
 
#code here
fig, ax = plt.subplots()
sns.despine(fig, left=True, bottom=True)
clarity_ranking = df['clarity'].sort_values()
sns.scatterplot(x='carat', y='price', hue='clarity',hue_order=clarity_ranking, size="depth", data=df, ax=ax, palette=palette, sizes=(1,8), linewidth=0)
plt.show()
```

# Soal 2: Catplot

Buatlah sebuah Categorical plot dengan kind = 'boxplot', sehingga diperoleh hasil seperti expected output berikut:


```python
#code here
sns.set_theme(style='ticks')
sns.catplot(data=df, x='clarity', y='price', kind='box', col='cut', hue='clarity', sharey=True, fliersize=0, legend=False)
plt.show()

```

# Soal 3: Pair Plot

Ambil 300 data secara acak dari df, kemudian buat pair plotnya dari 300 data tersebut untuk kolom carat, depth, table, dan price serta markers =  ['p', 'o', 's', 'D', 'X'] sehingga dihasilkan output sesuai expected.

```python
df = df.sample(n=300, random_state=123)

#code here
sns.set_theme(style='darkgrid')
sns.pairplot(df, vars=['carat','depth','table','price'], x_vars=['carat','depth','table','price'], y_vars=['depth','table','price'], markers=['p', 'o', 's', 'D', 'X'], hue='cut', height=2.5, corner=True)
plt.show()
```

# Soal 4: Joint Plot

Gunakan data yang sama dengan soal 3 untuk menghasilkan dua visualisasi joint plot antara carat dan price nya sesuai expected output.

```python
#code here
with sns.axes_style('white'):
    sns.jointplot(data=df, x='carat', y='price', kind='kde')
with sns.axes_style('darkgrid'):
    sns.jointplot(data=df, x='carat', y='price', kind='reg')
```
