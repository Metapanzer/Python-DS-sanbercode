**Soal 4. Time grouping**

 

*   Lakukan grouping data perubahan iklim pada tahun 2016 berdasarkan frekuensi 1 bulan dengan memberikan agregasi max pada kolom meantemp. humidity, wind_speed dan meanpressure!
*   Berikan analisa, di bulan apa rata-rata temperatur, humidity, wind_speed dan meanpressure tertinggi?





```python
#code here
group_2016_max = df["2016-01-01":"2016-12-31"].groupby(pd.Grouper(freq="M")).agg("max")
group_2016_max
```

Expected output:

(Gambar hanya sebagian)

![alt text](https://drive.google.com/uc?id=1s2Y8Dc3M3xbG-51mTuTmXUFgLvHDvc5Q)

```python
max_meantemp = group_2016_max["meantemp"] == group_2016_max["meantemp"].max()
max_humidity = group_2016_max["humidity"] == group_2016_max["humidity"].max()
max_wind_speed = group_2016_max["wind_speed"] == group_2016_max["wind_speed"].max()
max_meanpressure = group_2016_max["meanpressure"] == group_2016_max["meanpressure"].max()
display(group_2016_max[max_meantemp])
display(group_2016_max[max_humidity])
display(group_2016_max[max_wind_speed])
display(group_2016_max[max_meanpressure])
```

Jawab disini:Rata-rata meantemp tertinggi di bulan Mei, humidity tertinggi di bulan Desember, wind speed dan meanpressure tertinggi di bulan Maret
