<h2>Soal 1: Preparing Data</h2>

Manipulasi data tersebut dengan menggroupkan berdasarkan Genre, kemudian ambil rata2 penjualan untuk Global_Sales berdasarkan kategori Genre.

<h2>Soal 3: Preparing Data</h2>

Manipulasi data tersebut dengan menggroupkan berdasarkan Genre, kemudian ambil rata2 penjualan untuk setiap Region Kecuali Global_Sales berdasarkan kategori Genre.

```python
#code here
group_by_genre = df.groupby("Genre")[["NA_Sales", "EU_Sales", "JP_Sales", "Other_Sales"]].mean()
display(group_by_genre)
```

<h2>Soal 4: Perbandingan Kuantitatif Barplot - Grouping Visualisasi Dengan Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menghasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
import numpy as np
import seaborn as sns
group_trans = group_by_genre.transpose()

x = np.arange(len(group_trans.index))

fig, ax = plt.subplots(figsize=(14, 8))
colors = ["tab:red","tab:blue","tab:purple","tab:gray","tab:olive","tab:green","tab:pink","tab:orange","black","tab:brown","tab:pink","tab:cyan"]
width = 0.07

for index, genre in enumerate(group_trans.columns):
    ax.bar(x+(width*index), group_trans[genre], label=genre, width=width, color=colors[index])

ax.set_title("Mean Sales Video Games by Genre")
ax.set_ylabel("Mean Sales")
ax.set_xlabel("Region Sales")
sns.set_style("darkgrid")
plt.legend(loc="upper right", bbox_to_anchor=(1.06, 1.16))
plt.xticks(x + 0.4, labels=group_trans.index)
plt.show()
```

Genre Platform memiliki penjualan tertinggi di seluruh wilayah (NA, EU, JP, dan lainnya).\
Genre Role-Playing memiliki penjualan yang tinggi di Jepang (JP_Sales) dibandingkan dengan wilayah lainnya.\
Genre Shooter memiliki penjualan yang tinggi di North America dan Eropa.\
Adventure, Puzzle, dan Strategy memiliki penjualan yang relatif lebih rendah dibandingkan dengan genre lainnya di seluruh wilayah.
