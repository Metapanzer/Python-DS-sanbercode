<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(2*-np.pi, 2*np.pi, 200)

tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
#code here
fig, ax = plt.subplots(nrows=1, ncols=3,sharex=True,sharey=True, figsize=(12, 8))
data = [tan, cos, sin]
color = ['blue', 'red', 'green']
label= ['tanX', 'sinX', 'cosX']

for index, data in enumerate(data):
    ax[index].plot(x, data, color=color[index], label=label[index])
    ax[index].set_title(f'Grafik {label[index]}')
    
fig.legend(loc='upper center')
plt.axis([-6,6,-10,10])
plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
#code here
fig, ax = plt.subplots(figsize=(12, 8))
ax.plot(x, y, color='blue', label='cosX')
ax.set_title('Grafik CosX')

ins1 = ax.inset_axes([-0.1, -0.1, 0.3, 0.3])
ins1.plot(x, y, color='blue', label='cosX')
ins1.plot(x, y2, color='red', label='cosX^2')
ins1.set_title('Grafik CosX^2')

ins2 = ax.inset_axes([0.9, -0.1, 0.3, 0.3])
ins2.plot(x, y, color='blue', label='cosX')
ins2.plot(x, y3, color='green', label='cosX^3')
ins2.set_title('Grafik CosX^3')

ins3 = ax.inset_axes([0.9, 0.9, 0.3, 0.3])
ins3.plot(x, y, color='blue', label='cosX')
ins3.plot(x, y4, color='black', label='cosX^4')
ins3.set_title('Grafik CosX^4')

ins4 = ax.inset_axes([-0.1, 0.9, 0.3, 0.3])
ins4.plot(x, y, color='blue', label='cosX')
ins4.plot(x, y5, color='orange', label='cosX^5')
ins4.set_title('Grafik CosX^5')

plt.show()
```
